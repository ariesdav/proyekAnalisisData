### Cleaning Data

**Mengubah tipe data object menjadi datetime** pada kolom  kolom order_purchase_timestamp, order_approved_at, order_delivered_carrier_date, order_delivered_customer_date, dan order_estimated_delivery_date dalam **orders_df**

```python
datetime_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

for column in datetime_columns:
    orders_df[column] = pd.to_datetime(orders_df[column])

orders_df.info()
```

**Menambahkan nilai 'none'** pada kolom order_status.

```python
orders_df.fillna({'order_status': 'none'}, inplace=True)

orders_df.isna().sum()
```

Pengisian **missing value** pada kolom order_purchase_timestamp, order_approved_at, order_delivered_carrier_date, order_delivered_customer_date, dan order_estimated_delivery_date dengan **nilai default** yang diambil dari **tanggal terawal pada kolom order_purchase_timestamp**

```python
miss_orders_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

default_date = orders_df['order_purchase_timestamp'].min()

for column in miss_orders_columns:
    if column in orders_df.columns:
        orders_df[column].fillna(default_date, inplace=True)

orders_df.isna().sum()
```

**Mengubah tipe data dari object ke datetime** pada kolom shipping_limit_date dalam dataframe **order items**

```python
order_items_df['shipping_limit_date'] = pd.to_datetime(order_items_df['shipping_limit_date'])
order_items_df.info()
```

**Menambahkan value "unknown"** pada kolom product_category_name dan **value=0** pada kolom product_name_lenght, product_description_lenght, product_photos_qty

```python
products_df[products_df.product_category_name.isna()]
```

```python
products_df.fillna({
    'product_category_name': 'unknown',
    'product_name_lenght': 0,
    'product_description_lenght': 0,
    'product_photos_qty': 0
}, inplace=True)

products_df.isna().sum()
```

**Menambahkan value=0** pada kolom product_weight_g, product_length_cm, product_height_cm, product_width_cm.

```python
products_df[products_df.product_weight_g.isna()]
```

```python
miss_products_columns1 = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
products_df.fillna({column: 0 for column in miss_products_columns1}, inplace=True)

products_df.isna().sum()
```

**Insight:**
-  Pada orders_df, Tipe data datetime diubah untuk mempermudah analisis waktu dan menambahkan Nilai **'none'** ditambahkan untuk mengisi status pesanan yang kosong dan missing value pada kolom tipe data datetime dengan **nilai default dari tanggal terawal di order_purchase_timestamp**, sehingga tidak ada lagi nilai yang hilang.

- Pada order items, tipe data datetime diubah untuk kolom shipping_limit_date.

- Produk tanpa kategori diisi dengan **'unknown'** untuk menjaga konsistensi data. Kolom nama, deskripsi, dan jumlah foto produk diisi dengan **0** agar tidak ada nilai kosong. Berat, panjang, tinggi, dan lebar produk yang kosong juga diisi dengan **0** agar tidak ada NaN.

## Exploratory Data Analysis (EDA)

###Explore data orders_df




Membuat sebuah kolom baru **order_month** untuk memuat bulan berdasarkan order puschase_timestamp.

```python
orders_df['order_month'] = orders_df['order_purchase_timestamp'].dt.month
orders_df.sample(5)
```

Membuat sebuah kolom baru **season** untuk memuat musim berdasarkan order puschase_timestamp.

```python
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    elif month in [12, 1, 2]:
        return 'Winter'
    else:
        return 'Unknown'

orders_df['season'] = orders_df['order_month'].apply(get_season)

orders_df.sample(5)
```

###Explore data products_df dan product_category_name_eng_df

Melakukan merge pada dataframe products_df dengan product_category_name_eng_df

```python
product_categories_df = pd.merge(
    left=products_df,
    right=product_category_name_eng_df,
    how="left",
    right_on="product_category_name",
    left_on="product_category_name"
)

product_categories_df.head()
```

Bagian ini bertujuan untuk menghindari nilai NaN pada kolom product_category_name, karena tidak semua kategori produk dapat diubah ke dalam bahasa Inggris.

```python
product_categories_df['product_category_name_english'] = np.where(
    product_categories_df['product_category_name_english'].isnull(),
    product_categories_df['product_category_name'],
    product_categories_df['product_category_name_english']
)
```

###Explore data order_items_df dan orders_df

Menggabungkan dataframe order_items_df dengan orders_df

```python
orders_items_df = pd.merge(
    left=order_items_df,
    right=orders_df,
    how="left",
    right_on="order_id",
    left_on="order_id"
)

orders_items_df.head()
```

Menampilkan urutan jumlah order per musim

```python
orders_items_df.groupby(by='season').order_id.nunique().sort_values(ascending=False).reset_index()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Musim apakah pengguna sering melakukan pemesanan di e-comemrce ?


Menambahkan persentase order per musim menggunakan pivot table

```python
pivot_table_df = orders_items_df.pivot_table(
    index='season',
    values='order_id',
    aggfunc='nunique'
).reset_index()
total_orders = pivot_table_df['order_id'].sum()
pivot_table_df['percentage_orders'] = (pivot_table_df['order_id'] / total_orders) * 100
pivot_table_df.columns = ['season', 'orders', 'percentage_orders']
pivot_table_df = pivot_table_df.sort_values(by='percentage_orders', ascending=False).reset_index(drop=True)
print(pivot_table_df)
```

Menampilkan visualisasi pie chart

```python
plt.figure(figsize=(10,8))
plt.pie(
    pivot_table_df['orders'],
    labels=pivot_table_df['season'],
    autopct='%1.1f%%',
    startangle=100,
    textprops={'fontweight': 20}
)
plt.title('User Purchase Percentage by Season', fontsize=14)
plt.axis('equal')
plt.show()
```

### Pertanyaan 2: Kategori produk apa yang sering dibeli pengguna e-commerce berdasarkan musim?

Menampilkan kelompok data berdasarkan musim dan kategori produk

```python
season_product_category = main_data_df.groupby(by=["season", "product_category_name_english"]).order_id.nunique().reset_index()
print(season_product_category)
```

```python
top_season_product_categories = season_product_category.sort_values(by=["season", "order_id"], ascending=[True, False])
print(top_season_product_categories)
```


Menampilkan tabel untuk setiap musim dan memilih 5 kategori produk teratas dengan jumlah terbanyak di setiap musim.

```python
plt.figure(figsize=(16, 12))
unique_seasons = top_season_product_categories['season'].unique()

for i, season in enumerate(unique_seasons):
    plt.subplot(2, 2, i + 1)
    season_data = top_season_product_categories[top_season_product_categories['season']==season].head(5)
    sns.barplot(
        x='order_id',
        y='product_category_name_english',
        data=season_data,
        color='#1f77b4'
    )
    plt.title(f'Top 5 Product Categories in {season}', fontsize=14)
    plt.xlabel('Number of Orders', fontsize=12)
    plt.ylabel('Product Categories', fontsize=12)
    plt.xlim(0, season_data['order_id'].max() + 10)
    plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

**Insight:**
- Menyediakan pemahaman yang jelas mengenai proporsi pembelian menurut musim, **musim panas**lah menjadi favorit pelanggan, di mana persentase tersebut mencerminkan pola musim konsumen dalam memesan produk.

- Kategori bed_bath_table, health_beauty, dan sports_leisure mendominasi penjualan sepanjang tahun, menunjukkan peluang untuk meningkatkan penjualan melalui strategi pemasaran yang lebih fokus pada kategori ini di setiap musim.
